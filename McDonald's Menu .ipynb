### Plot which items contains which nutritions
### Item vs Nutrition

```python
# Let's define a function which can plot N Items which contains highest nutrition
# Note : Change value of N as per convenience in visualization 
```

```python
def plot_item_vs_nutrition(nutrition, n):
    item = menu.groupby(menu[nutrition])['Item'].sum().sort_index()[-n:]
    plt.figure(figsize=(9,10))
    sns.barplot(x=item.index, y=item.values)
    plt.show()
```

```python
# Let's Plot Sugar
plot_item_vs_nutrition('Sugars', 20)
```

- We can see Mostly 'Shakes' contains High Sugar

```python
# Let's plot Vitamin C (% Daily Value)
plot_item_vs_nutrition('Vitamin C (% Daily Value)', 5)
```

- We can see 'Minute Maid Orange Juice (Large)' contains highest % of Vitamin C

```python
# Let's plot Vitamin A (% Daily Value)
plot_item_vs_nutrition('Vitamin A (% Daily Value)', 5)
```

- By observing above graph we can see maximum % of Vitamin A presents in 'Salad'

```python
# Let's Plot Protein
plot_item_vs_nutrition('Protein', 20)
```

- Clearly, Chicken is best source of Protein

```python
# Let's Plot Cholesterol
plot_item_vs_nutrition('Cholesterol', 9)
```

### Plot Average Calories per Category

```python
avg_cal_per_cat = menu.groupby('Category')['Calories'].mean().reset_index()

plt.figure(figsize=(13, 8))
ax = sns.barplot(x='Category', y='Calories', data=avg_cal_per_cat)
plt.xticks(rotation=55, ha='right')
plt.title('Average Calories per Category')
for p in ax.patches:
    ax.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()
plt.show()
```

- Chicken & Fish contains at most average calories
- Overall Chicken & Fish, Smoothies & Shakes, Beef & Pork and Breakfast Contains higher Average Calories

### Whole Egg vs Egg White

```python
# Start get all Items which contains Egg White

egg_whites = menu[menu['Item'].str.contains('Egg White')]
```

```python
egg_whites_sort = egg_whites.sort_values(by='Calories')
sns.barplot(x='Calories', y='Item', data=egg_whites_sort, orient='h')
```

```python
# Get all Items which contains Whole Egg

whole_egg = menu[menu['Item'].str.contains('Egg')] 
whole_egg = whole_egg[~whole_egg['Item'].str.contains('White')]
```

```python
whole_egg_sort = whole_egg.sort_values(by='Calories')
sns.barplot(x='Calories', y='Item', data=whole_egg_sort, orient='h')
```

```python
# Average calories on Egg Whites Category
avg_whites_cal = egg_whites.Calories.mean(axis=0)
print("Average calories on Egg Whites category is ", round(avg_whites_cal, 2), "Cal.")

# Average calories on Whole Egg Category
avg_whole_cal = whole_egg.Calories.mean(axis=0)
print("Average calories on Whole Egg category is ", round(avg_whole_cal, 2), "Cal.")
```

- From analysis above, we can conclude that Items in Egg Whites category has higher average amount of calories than Items in Whole Eggs category.

### Plot Nutrition's Comparison
### Egg White Vs Whole Egg

```python
# Let's make a function for plotting comparison
# Fucntion takes list of nutrition - for better visualization make a list of 3 nutritions

def nutritions_comparison(n_list):
    barWidth = 0.25
    plt.figure(figsize=(10,5))

    r = np.arange(2)

    for n in n_list: 
        bar = [egg_whites[n].mean(axis=0), whole_egg[n].mean(axis=0)]
        plt.bar(r, bar, width=barWidth, edgecolor='white', label=n)
        r = [x + barWidth for x in r]
    plt.xticks([r + barWidth for r in range(2)], ['Egg White', 'Whole Egg'])
    plt.legend()
    plt.show()
```

```python
# Let's compare these 3 nutritions
n_list = ['Protein',
 'Calcium (% Daily Value)',
 'Iron (% Daily Value)']
nutritions_comparison(n_list)
```

- We Can See 'Whole Egg' contains higher amount of 'Calcium (% Daily Value)' & 'Iron (% Daily Value)'
- But 'Egg White' contains higher 'Protein' than 'Whole Egg' as expected.

```python
# Let's compare these 2 nutritions
n_list = ['Vitamin A (% Daily Value)',
 'Vitamin C (% Daily Value)']
nutritions_comparison(n_list)
```

- We can see 'Whole Egg' contains higher amount of 'Vitamin A (% Daily Value)' than 'Egg White' by visualizing comparison.
